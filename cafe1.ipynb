```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os 
import warnings
warnings.filterwarnings("ignore")
```

```python
coffe_domestic= pd.read_csv('data_cafe/Coffee_domestic_consumption.csv')
coffe_export= pd.read_csv('data_cafe/Coffee_export.csv')
coffe_import= pd.read_csv('data_cafe/Coffee_import.csv')
coffe_production= pd.read_csv('data_cafe/Coffee_production.csv')
```

```python
#Los paises que mas exportan cafe: 
mayor_exportador = coffe_export[['Country', 'Total_export']].sort_values(by=['Total_export'], ascending=False).head(10)

# Los paises que mas importan cafe:  
mayor_importador = coffe_import[['Country', 'Total_import']].sort_values(by=['Total_import'], ascending=False).head(10)

# Los pais que mas consumen cafe domesticamente:
mayor_consumidor = coffe_domestic[['Country', 'Total_domestic_consumption']]
top_consum = mayor_consumidor.sort_values(by=['Total_domestic_consumption'], ascending=False).head(10)

# Los paises que mas producen cafe: 
mayor_productor = coffe_production[['Country', 'Total_production']].sort_values(by='Total_production', ascending=False).head(10)
```

```python
# Plot principales exportadores y principales importadores
f, axs = plt.subplots(2, 1, figsize = (5,12))

sns.barplot(data = mayor_exportador, x='Total_export', y='Country', ax = axs[0], palette="hsv")
sns.barplot(data = mayor_importador, x='Total_import', y='Country', ax = axs[1],palette="hsv")

axs[0].set_title('Principales Exportadores')
axs[1].set_title('Principales Importadores')
```

```python
# Plot principales consumidores y principales productores
f, axs = plt.subplots(2, 1, figsize = (8,18))

sns.barplot(data = mayor_consumidor, x='Total_domestic_consumption', y='Country', ax = axs[0],palette="hsv")
sns.barplot(data = mayor_productor, x='Total_production', y='Country', ax = axs[1], palette="hsv")

axs[0].set_title('Principales Consumidores')
axs[1].set_title('Principales Productores')
```
