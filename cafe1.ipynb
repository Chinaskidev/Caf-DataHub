Este repositorio ofrece un análisis detallado sobre la industria del café a nivel mundial, con un enfoque especial en los principales países exportadores, importadores, consumidores y productores. Se presta particular atención a El Salvador, examinando datos desde 1990 hasta 2019 para ofrecer una visión clara de su posición en el mercado global. Desde la cantidad de café producido hasta las cifras de exportación y consumo, este proyecto proporciona una comprensión profunda de la participación del país en la industria cafetalera internacional.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os 
import warnings
warnings.filterwarnings("ignore")
```

```python
coffe_domestic= pd.read_csv('data_cafe/Coffee_domestic_consumption.csv')
coffe_export= pd.read_csv('data_cafe/Coffee_export.csv')
coffe_import= pd.read_csv('data_cafe/Coffee_import.csv')
coffe_production= pd.read_csv('data_cafe/Coffee_production.csv')
```

```python
#Los paises que mas exportan cafe: 
mayor_exportador = coffe_export[['Country', 'Total_export']].sort_values(by=['Total_export'], ascending=False).head(10)

# Los paises que mas importan cafe:  
mayor_importador = coffe_import[['Country', 'Total_import']].sort_values(by=['Total_import'], ascending=False).head(10)

# Los pais que mas consumen cafe domesticamente:
mayor_consumidor = coffe_domestic[['Country', 'Total_domestic_consumption']]
top_consum = mayor_consumidor.sort_values(by=['Total_domestic_consumption'], ascending=False).head(10)

# Los paises que mas producen cafe: 
mayor_productor = coffe_production[['Country', 'Total_production']].sort_values(by='Total_production', ascending=False).head(10)
```

```python
# Plot principales exportadores y principales importadores
f, axs = plt.subplots(2, 1, figsize = (5,12))

sns.barplot(data = mayor_exportador, x='Total_export', y='Country', ax = axs[0], palette="hsv")
sns.barplot(data = mayor_importador, x='Total_import', y='Country', ax = axs[1],palette="hsv")

axs[0].set_title('Principales Exportadores')
axs[1].set_title('Principales Importadores')
```

```python
# Plot principales consumidores y principales productores
f, axs = plt.subplots(2, 1, figsize = (8,18))

sns.barplot(data = mayor_consumidor, x='Total_domestic_consumption', y='Country', ax = axs[0],palette="hsv")
sns.barplot(data = mayor_productor, x='Total_production', y='Country', ax = axs[1], palette="hsv")

axs[0].set_title('Principales Consumidores')
axs[1].set_title('Mayor Productores')
```

EL SALVADOR <br>
Cuanto Produce, Exporta y Consume café. (Debo Aclarar que estos datos son desde 1990 hasta 2019)


```python
# Cuanto cafe exporta El Salvador?
# en estas lineas de codigo se puede ver la cantidad de cafe que exporta el salvador
pais = 'El Salvador'

datos_pais = coffe_export[coffe_export['Country'] == pais]

total_exportacion = datos_pais['Total_export'].sum()

#print(f'{pais} exporta {total_export} unidades de café.')
print(f'{pais} exporta {format(total_exportacion, ",")} unidades de café.') # este formato es para que se vea mas legible el numero
```

```python
# Cuanto cafe produce El Salvador?
# en estas lineas de codigo se puede ver la cantidad de cafe que produce el salvador.
pais = 'El Salvador'

datos_pais = coffe_production[coffe_production['Country'] == pais]

total_produccion = datos_pais['Total_production'].sum()

print(f'{pais} produce {format(total_produccion, ",")} unidades de café.')
```

```python
# Y cuanto cafe consume El Salvador?
# en estas lineas de codigo se puede ver la cantidad de cafe que consume el salvador.
pais = 'El Salvador'

datos_pais = coffe_domestic[coffe_domestic['Country'] == pais]

total_consumo = datos_pais['Total_domestic_consumption'].sum()

print(f'{pais} consume {format(total_consumo, ",")} unidades de café.')
```

```python
# Añadir una nueva columna de índice
mayor_exportador = coffe_export.reset_index(drop=True)

# Encontrar la posición de El Salvador
posicion = mayor_exportador[mayor_exportador['Country'] == 'El Salvador'].index[0] + 1

print(f'El Salvador es el {posicion}º mayor exportador de café.')
```

```python
fig, ax = plt.subplots(figsize = (8,18))

ax.barh(mayor_exportador['Country'], mayor_exportador['Total_export'], color='skyblue',edgecolor='black')

# ax.barh resalta la barra correspondiente a El Salvador
ax.barh('El Salvador', mayor_exportador.loc[mayor_exportador['Country'] == 'El Salvador', 'Total_export'], color='red', edgecolor='black')

# ax.invert_yaxis invierte el eje y para que el país con el mayor valor esté en la parte superior
ax.invert_yaxis()

ax.set_xlabel('Exportaciones Totales')
ax.set_title('Top Exportadores de Café')

plt.show()
```
